```python
#import dependencies
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.metrics import accuracy_score, balanced_accuracy_score, confusion_matrix 
from sklearn.model_selection import train_test_split, cross_val_score 
from imblearn.over_sampling import RandomOverSampler
from imblearn.combine import SMOTEENN
from imblearn.metrics import classification_report_imbalanced
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.inspection import permutation_importance
import seaborn as sns #used for KDE plots
from yellowbrick.classifier import ClassificationReport 
import matplotlib.pyplot as plt #used for histograms
import matplotlib.ticker as ticker
```

-------------------------------------------------------------------------------------------------------

Exploratory Data Analysis and Data Preprocessing
-----------------------------------------------------------------------------------------------------------
-----------------------------------------------------------------------------------------------------

```python
#read and view data
df = pd.read_csv("../cleaned_data_final.csv")

# Review the DataFrame
df
```

```python
#remove all rows that represent closed loans as reason for closure is unknown (prepayment risk, etc.)
df = df[df['mrtg_status'] != 500]
df
```


Number of rows reduced from 54,651 to 27,963


```python
#assign value of 0 to denote loans that are currently performing and 1 to denote loans that are delinquent/in default
df.loc[df['mrtg_status'] != 100, 'mrtg_status'] = 1
df.loc[df['mrtg_status'] == 100, 'mrtg_status'] = 0
```

```python
#drop fixed to float column/feature
df = df.drop('cd_fxfltr',axis=1)
df
```

```python
#rename columns for increased clarity to those unfamiliar with documentation (presentation purposes)
df = df.rename(columns={'cnt_mrtg_term': 'Mortgage Term',
                   'mrtg_status': 'Mortgage Status',
                   'rate_ltv': 'Original Loan to Value',
                   'cnt_blln_term': 'Balloon Term',
                   'cnt_io_per': 'Interest Only Period',
                   'cnt_rsdntl_unit': 'Count Residential Units',
                   'code_int': 'Fixed vs. Variable Interest Rate',
                   'rate_dcr':'Original Debt Service Coverage',
                   'rate_int':'Interest Rate'})

```

```python
#Generate correlation matrix & view data
correlation_matrix = df.corr()
correlation_matrix
```

---
Correlation Matrix Results
---
No variables appear to be highly correlated (<0.5). 

Most of the correlation coefficients are close to 0 or relatively small in magnitude, indicating weak or no linear relationship. There are a few coefficients that deviate from 0, but none of them appear to be close to 1 or -1, indicating a strong linear relationship.

Based on these observations, it seems there is no strong linear relationship between the features in the dataset. 
This may indicate that logistic regression is not the best modeling solution for our predictive analysis. We therefore expect decreased performance from our previous logistic regression models, which included a highly correlated feature, the Fixed to Float feature.

-----------------------------------------------------------------------------------------------------
MODEL ONE: Use RandomOverSampler and MinMax Scaler
----
----

```python
# Separate the y variable/target of "Mortgage Status"
y = df["Mortgage Status"]

# Separate the X variable, or the features
X = df.drop(columns="Mortgage Status")
```

```python
#correct data imbalance using RandomOverSampler from ImbalancedLearn
ros = RandomOverSampler(random_state=1)
X_resampled, y_resampled = ros.fit_resample(X, y)

#check balance of our target values once RandomOverSampler has been applied
print(y_resampled.value_counts())

#Split the data into train/test
X_res_train, X_res_test, y_res_train, y_res_test = train_test_split(X_resampled, 
                                                    y_resampled, 
                                                    random_state=35)
```

```python
#Rescale the data using MinMaxScaler

#Instantiate the scaler
scaler = MinMaxScaler()
#fit the scaler to training data
scaler.fit(X_res_train)
#transform training and test data
X_res_train_scaled = scaler.transform(X_res_train)
X_res_test_scaled = scaler.transform(X_res_test)
```

```python
# Instantiate the Logistic Regression model and assign a random_state parameter of 1 to the model
model1 = LogisticRegression(solver='lbfgs', max_iter=1500, random_state=1)

# Perform cross-validation
cv_scores = cross_val_score(model1, X_res_train, y_res_train, cv=5)

# Fit the model using the resampled training data
model1.fit(X_res_train_scaled, y_res_train)

# Get the list of feature names
feature_names = X.columns.tolist()

# Make a prediction using the testing data
predictions = model1.predict(X_res_test_scaled)

# Calculate the accuracy score
model1_as = accuracy_score(y_res_test, predictions)

# Calculate balanced accuracy score
model1_bas = balanced_accuracy_score(y_res_test, predictions)

# Create a confusion matrix
model1_cm = confusion_matrix(y_res_test, predictions)
# Plot the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(model1_cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Model 1: Confusion Matrix')
plt.show()

# Print classification report from Imbalanced Learn library
model1_icr = classification_report_imbalanced(y_res_test, predictions)
# Print classification report using yellowbrick
visualizer = ClassificationReport(model1, classes=['Performing', 'Delinquent'], title='Model 1: Classification Report')
visualizer.fit(X_res_train_scaled, y_res_train)
visualizer.score(X_res_test_scaled, y_res_test)
visualizer.show()


# Print the evaluation metrics
print("")
print("")
print("Model 1 Evaluation Metrics")
print("")
print("Cross-Validation Scores:", cv_scores)
print("Average Cross-Validation Score:", cv_scores.mean())
print("Accuracy:", model1_as)
print("Balanced Accuracy:", model1_bas)
print("")
print(model1_icr)
```
