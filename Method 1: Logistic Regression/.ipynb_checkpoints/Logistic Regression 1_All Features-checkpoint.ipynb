-------------------------------------------------------------------------------------------------------
MODEL SELECTION: 
-----------------------------------------------------------------------------------------------------
Logistic Regression
-----

Reason for the selection:
The goal of this project is to predict whether a loan is likely to perform well or default based on a selection of inputs. Logistic regression is specifically designed for binary classification tasks, designating it as a viable model to evaluate as a prediction method. Logistic regression provides interpretable results by estimating the probabilities of default based on the input features. Logistic regression models generate coefficients for each feature, indicating the direction and strength of their influence on the loan default probability. This interpretability can be useful for understanding the drivers of loan performance.

Logistic regression has certain limitations that must be considered. For example, it assumes a linear relationship between the features and the log-odds of the target variable, which may not always hold true in complex loan performance scenarios. Additionally, logistic regression may struggle with highly imbalanced datasets or situations where non-linear relationships are prominent. In such cases, more advanced techniques like ensemble methods or nonlinear models may be more appropriate.

```python
#read and view data
df = pd.read_csv("../cleaned_data_final.csv")

# Review the DataFrame
df
```

```python
#Initial check of distribution of features

#Create histograms for each feature in dataset to check distribution

# Select the features
features = ['Mortgage Term',
            'Original Loan to Value',
            'Fixed to Float',
            'Balloon Term',
            'Interest Only Period',
            'Count Residential Units',
            'Fixed vs. Variable Interest Rate',
            'Original Debt Service Coverage',
            'Interest Rate']

# Create subplots for each feature
fig, axes = plt.subplots(nrows=1, ncols=len(features), figsize=(20, 4))

# Iterate over each feature and create a histogram
for i, feature in enumerate(features):
    ax = axes[i]
    ax.hist(df[feature], bins=20)
    ax.set_xlabel(feature)
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#Plot histograms by KDE plots for each feature (KDE plots will illustrate distribution more precisely)

# Select the features
features = ['Mortgage Term',
            'Original Loan to Value',
            'Fixed to Float',
            'Balloon Term',
            'Interest Only Period',
            'Count Residential Units',
            'Fixed vs. Variable Interest Rate',
            'Original Debt Service Coverage',
            'Interest Rate']

# Determine the number of rows and columns for subplots
nrows = len(features)
ncols = 2  # Two columns: Histogram and KDE

# Create subplots for each feature
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(12, 30))

# Iterate over each feature and create histograms and KDE plots
for i, feature in enumerate(features):
    hist_ax = axes[i, 0]
    kde_ax = axes[i, 1]
    
    # Histogram
    hist_ax.hist(df[feature], bins=20)
    hist_ax.set_xlabel(feature)
    hist_ax.set_ylabel('Frequency')
    
    # KDE plot
    sns.kdeplot(data=df, x=feature, ax=kde_ax)
    kde_ax.set_xlabel(feature)
    kde_ax.set_ylabel('Density')

plt.tight_layout()
plt.show()
```

---
IMBALANCED DATA SUMMARY:
---
Checking the balance of our target variable, mortgage status, revealed great imbalance within the dataset. That is, it contains almost 28 thousand instances of performing loans versus just over 200 loans in some form of default. Given this disparity, the data will need to be rebalanced prior to training and testing the model. Several methods, including RandomOverSampler from Imbalanced Learn and SMOTE-ENN (Synthetic Minority Over-Sampling Technique), will be used to determine which method performs best.

-----------------------------------------------------------------------------------------------------
MODEL ONE: Use RandomOverSampler and MinMax Scaler
----
----

```python
#correct data imbalance using RandomOverSampler from ImbalancedLearn
ros = RandomOverSampler(random_state=1)
X_resampled, y_resampled = ros.fit_resample(X, y)

#check balance of our target values once RandomOverSampler has been applied
print(y_resampled.value_counts())

#Split the data into train/test
X_res_train, X_res_test, y_res_train, y_res_test = train_test_split(X_resampled, 
                                                    y_resampled, 
                                                    random_state=35)
```

```python
#Rescale the data using MinMaxScaler

#Instantiate the scaler
scaler = MinMaxScaler()
#fit the scaler to training data
scaler.fit(X_res_train)
#transform training and test data
X_res_train_scaled = scaler.transform(X_res_train)
X_res_test_scaled = scaler.transform(X_res_test)
```

```python
# Instantiate the Logistic Regression model and assign a random_state parameter of 1 to the model
model1 = LogisticRegression(solver='lbfgs', max_iter=1500, random_state=1)

# Perform cross-validation
cv_scores = cross_val_score(model1, X_res_train, y_res_train, cv=5)

# Fit the model using the resampled training data
model1.fit(X_res_train_scaled, y_res_train)

# Get the list of feature names
feature_names = X.columns.tolist()

print("")
print("Model 1: Coefficient Importance with Corresponding Feature:")
# Print coefficient importance with corresponding feature names
coefficient_model1 = model1.coef_[0]
coefficient_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficient_model1})
coefficient_df = coefficient_df.sort_values(by='Coefficient', ascending=False)
print(coefficient_df.to_string(index=False))

print("")
print("Model 1: Permutation Importance with Corresponding Feature:")
# #View permutation importance
fi_model1 = permutation_importance(model1, X_res_train_scaled, y_res_train)
# Print permutation importance with corresponding feature names
importances = fi_model1['importances_mean']
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
print(importance_df.to_string(index=False))

# Visualize coefficient importance
plt.figure(figsize=(10, 6))
plt.barh(coefficient_df['Feature'], coefficient_df['Coefficient'])
plt.xlabel('Coefficient Importance')
plt.ylabel('Feature')
plt.title('Model 1: Coefficient Importance across Features')
plt.show()

# Visualize permutation importance
plt.figure(figsize=(10, 6))
plt.barh(importance_df['Feature'], importance_df['Importance'])
plt.xlabel('Permutation Importance')
plt.ylabel('Feature')
plt.title('Model 1: Permutation Importance Across Features')
plt.show()



# Make a prediction using the testing data
predictions = model1.predict(X_res_test_scaled)

# Calculate the accuracy score
model1_as = accuracy_score(y_res_test, predictions)

# Calculate balanced accuracy score
model1_bas = balanced_accuracy_score(y_res_test, predictions)

# Create a confusion matrix
model1_cm = confusion_matrix(y_res_test, predictions)
# Plot the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(model1_cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Model 1: Confusion Matrix')
plt.show()

# Print classification report from Imbalanced Learn library
model1_icr = classification_report_imbalanced(y_res_test, predictions)
# Print classification report using yellowbrick
visualizer = ClassificationReport(model1, classes=['Performing', 'Delinquent'], title='Model 1: Classification Report')
visualizer.fit(X_res_train_scaled, y_res_train)
visualizer.score(X_res_test_scaled, y_res_test)
visualizer.show()


# Print the evaluation metrics
print("")
print("")
print("Model 1 Evaluation Metrics")
print("")
print("Cross-Validation Scores:", cv_scores)
print("Average Cross-Validation Score:", cv_scores.mean())
print("Accuracy:", model1_as)
print("Balanced Accuracy:", model1_bas)
print("")
print(model1_icr)
```

---

MODEL ONE PERFORMANCE SUMMARY:
--
Note- This model was run multiple times while scaling up the number of iterations until it reached convergence.

For performing loans ('0'), the model correctly predicted  75.4% of its predictions (precision), caught 76.5% of positive cases (recall), and 76% (F-1) of its positive predicitions were correct. According to the support metric, the dataset included 6,965 occurences of performing loans. 

For non-performing loans ('1'), the model correctly predicted 75.9% of its predictions (precision), caught 74.8% of positive cases (recall), and 75.4% of positive predicitions were correct (F-1). According to the support metric, the dataset included 6,898 occurences of non-performing loans.

Based on the coefficient importance, the "Interest Rate" feature has the highest positive impact on the target variable, followed by "Fixed to Float" and "Original Loan to Value." Conversely, "Count Residential Units," "Mortgage Term," and "Balloon Term" have the highest negative impact on the target variable.

According to permutation importance, "Fixed to Float" is the most important feature, followed by "Interest Rate" and "Mortgage Term.""Fixed vs. Variable Interest Rate" and "Original Debt Service Coverage" have very low importance based on permutation importance.

The coefficient importance and permutation importance measures differ significantly, which could indicate the presence of multicollinearity or non-linear relationships between features and the target variable.

Overall, Model 1 demonstrates decent performance with a relatively high average cross-validation score (74.56%), indicating consistent performance across different data subsets. The accuracy (75.66%) and balanced accuracy (75.66%) scores also suggest that the model is reasonably effective in predicting the target variable, considering the class imbalances in the dataset.

```python
# Instantiate the Logistic Regression model and assign a random_state parameter of 1 to the model
model4 = LogisticRegression(solver='lbfgs', max_iter=1500, random_state=1)

# Perform cross-validation
cv_scores = cross_val_score(model4, X_res_train, y_res_train, cv=5)

# Fit the model using the resampled training data
model4.fit(X_res_train_scaled, y_res_train)

# Get the list of feature names
feature_names = X.columns.tolist()

print("")
print("Model 4: Coefficient Importance with Corresponding Feature:")
# Print coefficient importance with corresponding feature names
coefficient_model4 = model4.coef_[0]
coefficient_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficient_model4})
coefficient_df = coefficient_df.sort_values(by='Coefficient', ascending=False)
print(coefficient_df.to_string(index=False))

print("")
print("Model 4: Permutation Importance with Corresponding Feature:")
# #View permutation importance
fi_model4 = permutation_importance(model4, X_res_train_scaled, y_res_train)
# Print permutation importance with corresponding feature names
importances = fi_model4['importances_mean']
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
print(importance_df.to_string(index=False))

# Visualize coefficient importance
plt.figure(figsize=(10, 6))
plt.barh(coefficient_df['Feature'], coefficient_df['Coefficient'])
plt.xlabel('Coefficient Importance')
plt.ylabel('Feature')
plt.title('Model 4: Coefficient Importance across Features')
plt.show()

# Visualize permutation importance
plt.figure(figsize=(10, 6))
plt.barh(importance_df['Feature'], importance_df['Importance'])
plt.xlabel('Permutation Importance')
plt.ylabel('Feature')
plt.title('Model 4: Permutation Importance Across Features')
plt.show()



# Make a prediction using the testing data
predictions = model4.predict(X_res_test_scaled)

# Calculate the accuracy score
model4_as = accuracy_score(y_res_test, predictions)

# Calculate balanced accuracy score
model4_bas = balanced_accuracy_score(y_res_test, predictions)

# Create a confusion matrix
model4_cm = confusion_matrix(y_res_test, predictions)
# Plot the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(model4_cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Model 4: Confusion Matrix')
plt.show()

# Print classification report from Imbalanced Learn library
model4_icr = classification_report_imbalanced(y_res_test, predictions)
# Print classification report using yellowbrick
visualizer = ClassificationReport(model4, classes=['Performing', 'Delinquent'], title='Model 4: Classification Report')
visualizer.fit(X_res_train_scaled, y_res_train)
visualizer.score(X_res_test_scaled, y_res_test)
visualizer.show()

# Print the evaluation metrics
print("")
print("")
print("Model 4 Evaluation Metrics")
print("")
print("Cross-Validation Scores:", cv_scores)
print("Average Cross-Validation Score:", cv_scores.mean())
print("Accuracy:", model4_as)
print("Balanced Accuracy:", model4_bas)
print("")
print(model4_icr)
```
