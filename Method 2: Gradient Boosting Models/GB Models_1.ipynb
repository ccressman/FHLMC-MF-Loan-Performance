MODEL SELECTION: 
-----------------------------------------------------------------------------------------------------
Histogram-Based Gradient Boosting Model
-----

Reason for the selection:
Data analysis indicates that our data is not strongly correlated and that the features do not maintain a strong linear relationship. This is underlined by the relatively poor performance of our logistic regression models, which represent a nonoptimal method to build a predicitive model for this dataset. A more complex model, like Gradient Boosting Classification (GB), that is able to account for non-linear relationships is therefore a more appropriate model. 

Due to the large size of the dataset being evaluated (>50,000 rows), a histogram-based gradient boosting model was selected over a gradient boosting classifier. Model training is accelerated with this modeling technique by its use of discretizing, or binning, the input features to a few hundred unique values. 

Gradient boosting models are a type of ensemble machine learning (bagging, boosting, and stacking) that constructs an additive model so that predicition errors in the previous model are addressed in the subsequent model created.

Analysis of feature distribution indicates highly skewed data. Although Histogram-based gradient boosting models, do not typically require explicit data scaling, since the dataset's features have significantly different scales, they will be normalized to help prevent features with larger magnitudes from dominating the learning process.

IMBALANCED DATA SUMMARY:
---
Checking the balance of our target variable, mortgage status, revealed great imbalance within the dataset. That is, it contains almost 28 thousand instances of performing loans versus just over 200 loans in some form of default. Given this disparity, the data will need to be rebalanced prior to training and testing the model. Several methods, specifically RandomOverSampler from Imbalanced Learn and SMOTE-ENN (Synthetic Minority Over-Sampling Technique), will be used to determine which method performs best.


-----------------------------------------------------------------------------------------------------
MODEL ONE: Use RandomOverSampler and MinMax Scaler
----
----

```python
#correct data imbalance using RandomOverSampler from ImbalancedLearn
ros = RandomOverSampler(random_state=1)
X_resampled, y_resampled = ros.fit_resample(X, y)
```

```python
#check balance of our target values once RandomOverSampler has been applied
y_resampled.value_counts()
```

```python
#Split the data into train/test
X_res_train, X_res_test, y_res_train, y_res_test = train_test_split(X_resampled, 
                                                    y_resampled, 
                                                    random_state=35)
```

```python
#Rescale the data using MinMaxScaler

#instantiate the scaler
scaler = MinMaxScaler()
#fit the scaler to training data
scaler.fit(X_res_train)
#transform training and test data
X_res_train_scaled = scaler.transform(X_res_train)
X_res_test_scaled = scaler.transform(X_res_test)
```

```python
#Instantiate the Histogram based GBC model and fit it on resampled/rebalanced data
#using standard 255 bins and max iterations of 100
model1 = HistGradientBoostingClassifier(max_iter=100).fit(X_res_train_scaled, y_res_train)


# Perform cross-validation
cv_scores = cross_val_score(model1, X_res_train, y_res_train, cv=5)


# Get the list of feature names
feature_names = X.columns.tolist()


print("")
print("Model 1: Permutation Importance with Corresponding Feature:")
# #View permutation importance
fi_model1 = permutation_importance(model1, X_res_train_scaled, y_res_train)
# Print permutation importance with corresponding feature names
importances = fi_model1['importances_mean']
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
print(importance_df.to_string(index=False))

# Visualize permutation importance
plt.figure(figsize=(10, 6))
plt.barh(importance_df['Feature'], importance_df['Importance'])
plt.xlabel('Permutation Importance')
plt.ylabel('Feature')
plt.title('Model 1: Permutation Importance Across Features')
plt.show()


# Make a prediction using the testing data
predictions = model1.predict(X_res_test_scaled)

# Calculate the accuracy score
model1_as = accuracy_score(y_res_test, predictions)

# Calculate balanced accuracy score
model1_bas = balanced_accuracy_score(y_res_test, predictions)

# Create a confusion matrix
model1_cm = confusion_matrix(y_res_test, predictions)
# Plot the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(model1_cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Model 1: Confusion Matrix')
plt.show()

# Print classification report from Imbalanced Learn library
model1_icr = classification_report_imbalanced(y_res_test, predictions)
# Print classification report using yellowbrick
visualizer = ClassificationReport(model1, classes=['Performing', 'Delinquent'], title='Model 1: Classification Report')
visualizer.fit(X_res_train_scaled, y_res_train)
visualizer.score(X_res_test_scaled, y_res_test)
visualizer.show()


# Print the evaluation metrics
print("")
print("")
print("Model 1 Evaluation Metrics")
print("")
print("Cross-Validation Scores:", cv_scores)
print("Average Cross-Validation Score:", cv_scores.mean())
print("Accuracy:", model1_as)
print("Balanced Accuracy:", model1_bas)
print("")
print(model1_icr)
```

MODEL ONE PERFORMANCE SUMMARY:
--
For performing loans ('0'), the model correctly predicted  100% of its predictions (precision), caught 98% of positive cases (recall), and 99% (F-1) of its positive predicitions were correct. According to the support metric, the dataset included 6,965 occurences of healthy loans. 

For delinquent loans ('1'), the model correctly predicted 98% of its predictions (precision), caught 100% of positive cases (recall), and 99% of positive predicitions were correct (F-1). According to the support metric, the dataset included 6,898 occurences of healthy loans (siginificantly less than healthy loans).

According to permutation importance, "Interest Rate" is the most important feature, at 0.28, followed by "Fixed to Float", calculated at 0.26. This score indicates that these features are the most influential in predicting performance as calculated by this model.

---

Overall this model performs very well and minimizes type 2 error when predicting loan default.  

It reports a high average cross-validation score (99%), indicating consistent performance across different data subsets. The accuracy (98.97%) and balanced accuracy (98.98%) scores indicate that the model is highly effective in predicting the target variable.
